### Excercise 1
#### Write a function that takes OfferDetails as input, and returns FaceValue as output

```python
def get_FaceValue(string_S):                    # Taking  String as Input 
    for x in (string_S.split(' ')):                # Spliting the string by ' ' space between each wrods
        if ('$' in x) or ('¢'in x):             # Finding the element which contain dollar or cent symbol 
            return x                            # Return the Fave Value
```

```python
# Check the function functionality 
print(get_FaceValue(df[5:6].to_string()))
```

#### Run the function against all rows in the attached data and determine function accuracy (Higher the better)

```python
for x in range(0,df.size+1):
    if get_FaceValue(df[:x].to_string()):
        print(get_FaceValue(df[x:x+1].to_string()))
    else:
        print('No Value')
```
